```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('/Order Details.csv')
```

**Visualisasi Data**

```python
# Buat visualisasi analisis profit berdasarkan kategori dan sub-kategori
sns.set(style="darkgrid")
plt.figure(figsize=(10,6))
sns.barplot(x='Sub-Category', y='Profit', hue='Category', data=data)
plt.title('Analisis Profit Berdasarkan Kategori dan Sub-Kategori',fontsize=16, fontweight='bold')
plt.xlabel('Sub-Category', fontsize=12, fontweight='bold')
plt.ylabel('Profit', fontsize=12, fontweight='bold')
plt.grid(True, which='major', linestyle='--', linewidth=0.5)
plt.xticks(rotation=90)
sns.despine()
plt.show()
```
